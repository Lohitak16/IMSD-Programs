```python
1(B)
<!DOCTYPE html>
<html>
<head>
<title>String Object Example</title>
<script>
function showStringMethods() {
  let text = "Hello World!";
  document.write("Original String: " + text + "<br>");
  document.write("Length: " + text.length + "<br>");
  document.write("Uppercase: " + text.toUpperCase() + "<br>");
  document.write("Substring (0,5): " + text.substring(0,5));
}
</script>
</head>
<body onload="showStringMethods()">
</body>
</html>
```

```python
3(A)
<!DOCTYPE html>
<html>
<head>
<title>Voter Info</title>
<script>
function getVoterInfo() {
  let name = prompt("Enter your name:");
  let age = prompt("Enter your age:");
  let status = (age >= 18) ? "Eligible to Vote" : "Not Eligible to Vote";
  document.write("<table border='1'><tr><th>Name</th><th>Age</th><th>Status</th></tr>");
  document.write("<tr><td>" + name + "</td><td>" + age + "</td><td>" + status + "</td></tr></table>");
}
</script>
</head>
<body onload="getVoterInfo()">
</body>
</html>
```
